```python

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
simulation = pd.read_csv("simulations/simulation_4_3_update_lag_xgbclassifier2.pkl.csv")
plt.figure(figsize=(10, 5))

ml_model = simulation[simulation["model_type"]==1]
control_model = simulation[simulation["model_type"]==0]

print(f"Average change in ML PnL: {sum(ml_model["pnl"].diff().dropna()) / (len(ml_model)-1)}")
print(f"Average change in AS PnL: {sum(control_model["pnl"].diff().dropna()) / (len(control_model)-1)}")

plt.scatter(ml_model["sequence_number"], ml_model["pnl"], label="ML", color='blue', linewidth=2)
plt.plot(control_model["sequence_number"], control_model["pnl"], label="Cont", color='red', linewidth=2)


plt.xlabel("Sequence Number")
plt.ylabel("Change in Value ($)")
plt.title("PNL")
plt.legend()
plt.grid(True)
plt.show()
```
